```python
from time import time
import logging
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split    
from sklearn.datasets import fetch_lfw_people   #sk-learn中的人脸数据集
from sklearn.model_selection import GridSearchCV #调试函数的参数
from sklearn.metrics import classification_report #显示报告，显示主要的分类指标，返回每个类别的准确率，召回率等
from sklearn.metrics import confusion_matrix #真实类别和预测类别做出判断，以矩阵形式表示
from sklearn.decomposition import PCA   #主成分分析
from sklearn.svm import SVC   #支持向量机
from sklearn.cluster.tests.test_k_means import n_samples   
```

```python
lfw_people = fetch_lfw_people(min_faces_per_person=70,resize=0.4)
#每人保留70张，调整每张人脸大小比率为0.4
```

```python
n_samples,h,w = lfw_people.images.shape
print(n_samples,h,w)
```

```python
x = lfw_people.data
print(x.shape)
```

```python
n_features = x.shape[1]  #特征数
```

```python
y = lfw_people.target
print(y,max(y),min(y))   #标签范围是0-6
target_names = lfw_people.target_names
print(target_names)   #目标名称
```

```python
n_classes = target_names.shape[0]
```

```python
print('total dataet size:')
print('n_sample: %d' % n_samples)
print('n_features: %d' % n_features)
print('n_class: %d' % n_classes)
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)   #划分训练集和验证集
```

```python
n_components = 150
```

加入特征提取，将最初的1850个特征提取出150个主成分

```python
t0 = time()  #时间函数，计算运行时间
pca = PCA(n_components = n_components,whiten = True).fit(x_train)  #主成分分析
#n_components是要保留的成份数
print("done in %0.3fs" %(time() - t0))
eigenfaces = pca.components_.reshape((n_components,h,w))#人脸转化为特征向量集
```

```python
print("projecting the input data on the eigenfaces orthonomal basis")
t0=time()
x_train_pca=pca.transform(x_train)     #将训练数据进行降维
x_test_pca=pca.transform(x_test)     #将测试数据进行降维
print("done in %0.3fs " % (time()-t0))
```

```python
print("fitting the classfier to the training set ")
t0 = time()
parag_grid ={'C':[1e3,5e3,1e4,5e4,1e5],'gamma':[0.0001,0.0005,0.001,0.005,0.01,0.1],}
#C是正则化参数，
clf = GridSearchCV(SVC(kernel='rbf'),parag_grid)   #支持向量机分类
#rbf表示高斯核，gamma表示核函数的系数，以上述参数选择构造分类器
clf = clf.fit(x_train_pca,y_train)    #进行拟合
print("done in %0.3fs" %(time()-t0))
```

```python
print("best estimator found by grid search:")
print(clf.best_estimator_)   #挑选出最好的参数
```

```python
print("predict the people's name on the test set")
t0=time()
y_pred = clf.predict(x_test_pca)   #用训练后的分类器进行预测
print("done in %0.3fs " %(time()-t0))
print(classification_report(y_test,y_pred,target_names=target_names))
print(confusion_matrix(y_test,y_pred,labels=range(n_classes)))  #得到结果矩阵
```

上述结果解读：关于矩阵，类别为1的对象被分类为类别1的一共14个，分类为类别2的有2个，...。预测结果最好的是第4类，只有一个判断错误，被分类为第2类

```python
def plot_gallery(images,titles,h,w,n_row=3,n_col=4):
    plt.figure(figsize=(1.8*n_col,2.4*n_row))
    plt.subplots_adjust(bottom=0,left=.01,right=.99,top=.90,hspace=.35)
    for i in range(n_row*n_col):
        plt.subplot(n_row,n_col,i+1)
        plt.imshow(images[i].reshape((h,w)),cmap=plt.cm.gray)
        plt.title(titles[i],size=12)
        plt.xticks()
        plt.yticks()
```

```python
def title(y_pred,y_test,target_names,i):
    pred_name = target_names[y_pred[i]].rsplit(' ',1)[-1]  #从结果中取出真实名字和预测名字
    true_name = target_names[y_test[i]].rsplit(' ',1)[-1]
    return 'predicted :%s \ntrue:  %s' %(pred_name,true_name)
```

```python
prediction_titles=[title(y_pred,y_test,target_names,i)
                   for i in range(y_pred.shape[0])]
plot_gallery(x_test, prediction_titles, h, w)  #画出完整数据下的图像
```

```python
eigenface_title = ["eigenface %d" % i for i in range(eigenfaces.shape[0])]
plot_gallery(eigenfaces, eigenface_title, h, w)  #画出特征提取之后的图像
plt.show()
```
